Implement a function trapezoidal that approximates the integral of the function f on the domain from a to b using N points.

```python
def trapezoidal(f,a,b,N):
    # trap rule:
    # 1/(N-1)*[fx_1 + fx_2 .... fx_(n-1)]+1/2(N-1)*(fx_0 + fx_n)
    
    h = (b-a)/(N-1)    
    x = np.linspace(a,b,N)    
    f_x = np.array(eval(f))    
    area = (h/2)*(f_x[0]+ f_x[-1] + 2*np.sum(f_x[1:-1]))    
    return area
    
trapezoidal("x**2",0,10,10)
```
